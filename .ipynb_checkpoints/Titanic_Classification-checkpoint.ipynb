**IMPORT LIBRARIES**

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
import warnings 
warnings.simplefilter('ignore')
```

**LOADING DATA**

```python
df = pd.read_csv('Titanic.csv')
```

```python
df.head()
```

**EXPLORATORY DATA ANALYSIS**

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how = 'any')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df['PassengerId'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
df['Name'].value_counts()
```

```python
df = df.drop('Name',axis=1)
df = df.drop('Ticket',axis = 1)
df = df.drop('Cabin',axis=1)
```

```python
df.head()
```

```python
dummies = pd.get_dummies(df,columns=['Sex','Embarked'])
```

```python
df = pd.concat([df,dummies],axis=1)
```

```python
df = df.drop(['Sex','Embarked'],axis = 1)
```

```python
df.head()
```

```python
df.columns
```

**VISUALIZATION**

**MODEL BUILDING**

```python
x = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3 , random_state = 1,shuffle=True)
```

**LOGISTIC REGRESSION**

```python
from sklearn import metrics 

from sklearn.linear_model import LogisticRegression

# Fit the model on train
model = LogisticRegression(solver="liblinear")
model.fit(x_train,y_train)
```

```python
lr_predict = model.predict(x_test)
lr_predict
```

```python
lr_train_score = model.score(x_train,y_train)
lr_train_score
```

```python
lr_test_score = model.score(x_test,y_test)
lr_test_score
```

**DECISION TREE CLASSIFIER**

```python
from sklearn.tree import DecisionTreeClassifier
dTree = DecisionTreeClassifier(criterion='gini',random_state=1)
dTree.fit(x_train,y_train)
```

```python
dTree_predict = dTree.predict(x_test)
```

```python
dt_train_score = dTree.score(x_train,y_train)
dt_train_score
```

```python
dt_test_score = dTree.score(x_test,y_test)
dt_test_score
```

**BAGGING CLASSIFIER**

```python
from sklearn.ensemble import BaggingClassifier

bgcl = BaggingClassifier(base_estimator=dTree,n_estimators = 50,random_state =1)

bgcl = bgcl.fit(x_train,y_train)
```

```python
bgcl_predict = bgcl.predict(x_test)
```

```python
bgcl_train_score = bgcl.score(x_train,y_train)
bgcl_train_score
```

```python
bgcl_test_score = bgcl.score(x_test,y_test)
bgcl_test_score
```

**ADA BOOST**

```python
from sklearn.ensemble import AdaBoostClassifier
abcl = AdaBoostClassifier(base_estimator=dTree,n_estimators=10,random_state=1)
abcl = abcl.fit(x_train,y_train)
```

```python
abcl_predict = abcl.predict(x_test)
```

```python
abcl_train_score = abcl.score(x_train,y_train)
abcl_train_score
```

```python
abcl_test_score = abcl.score(x_test,y_test)
abcl_test_score
```

**GRADIENT BOOST CLASSIFIER**

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcl = GradientBoostingClassifier(n_estimators=50,random_state=1)
gbcl = gbcl.fit(x_train,y_train)
```

```python
gbcl_predict = gbcl.predict(x_test)
```

```python
gbcl_train_score = gbcl.score(x_train,y_train)
gbcl_train_score
```

```python
gbcl_test_score = gbcl.score(x_test,y_test)
gbcl_test_score
```

**XGB CLASSIFIER**

```python
from xgboost import XGBClassifier
xgb = XGBClassifier() 
xgb.fit(x_train, y_train)
```

```python
xgb_predict = xgb.predict(x_test)
```
